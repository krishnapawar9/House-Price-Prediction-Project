```python
df = pd.read_csv("House Cleaned Data.csv")
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
rfe = RFE(lm,)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

```python
# Create a DataFrame with the selected features
df_selected_features = X_train[col]

# Add the target variable 'price' back to the DataFrame for the heatmap
df_selected_features['price'] = y_train

# Plot the correlation heatmap
plt.figure(figsize = (10, 8))
sns.heatmap(df_selected_features.corr(), annot = True, cmap="YlGnBu")
plt.title('Correlation Heatmap of Selected Features')
plt.show()

# Drop the 'price' column from the selected features DataFrame to revert to the original state
df_selected_features.drop('price', axis=1, inplace=True)
```

```python
import statsmodels.api as sm
```

```python
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
# Create a DataFrame for the model coefficients
model_coefficients = pd.DataFrame(lm.params, columns=['Coefficient'])
model_coefficients = model_coefficients.drop('const')
model_coefficients['Feature'] = model_coefficients.index

# Sort the coefficients by their absolute value for better visualization
model_coefficients['Abs_Coefficient'] = model_coefficients['Coefficient'].abs()
model_coefficients = model_coefficients.sort_values(by='Abs_Coefficient', ascending=False)

# Plot the coefficients
plt.figure(figsize=(12, 8))
sns.barplot(x='Coefficient', y='Feature', data=model_coefficients, palette='viridis')
plt.title('Feature Coefficients of the Final Model')
plt.xlabel('Coefficient Value')
plt.ylabel('Features')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
y_train_price = lm.predict(X_train_rfe)
```

```python
res = (y_train_price - y_train)
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)
```

```python
num_vars = ['area','stories', 'bathrooms', 'airconditioning', 'prefarea','parking','price']
```

```python
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
```

```python
y_test = df_test.pop('price')
X_test = df_test
```

```python
X_test = sm.add_constant(X_test)
```

```python
X_test_rfe = X_test[X_train_rfe.columns]
```

```python
y_pred = lm.predict(X_test_rfe)
```

```python
# Calculate residuals for the test data
test_residuals = y_test - y_pred

# Distribution plot of test residuals
plt.figure(figsize=(10, 6))
sns.histplot(test_residuals, kde=True, bins=30)
plt.title('Distribution of Test Residuals')
plt.xlabel('Residuals')
plt.show()

# Scatter plot of predicted values vs test residuals
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, test_residuals, alpha=0.5)
plt.title('Predicted Values vs Test Residuals')
plt.xlabel('Predicted Prices')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test, y_pred)
```
